```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = ["CarCount", "BikeCount", "BusCount", "TruckCount", ]
values = [10, 20, 23, 30,]

plt.bar(labels, values)
plt.xlabel("Labels")
plt.ylabel("Count")
plt.title("Bar Chart of Labels")
plt.show()
```

```python
labels = ["CarCount", "BikeCount", "BusCount", "TruckCount", ]
sizes = [10, 20, 23, 30,]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart of Labels")
plt.show()
```

```python
import seaborn as sns

labels = ["CarCount", "BikeCount", "BusCount", "TruckCount", ]
data = [10, 20, 23, 30,]

sns.boxplot(x=labels, y=data)
plt.xlabel("Labels")
plt.ylabel("Values")
plt.title("Box Plot of Labels")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["CarCount", "BikeCount", "BusCount", "TruckCount", ]
data = [10, 20, 23, 30,]

for i in range(len(labels)):
    plt.hist(data[i], alpha=0.5, label=labels[i])

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Labels")
plt.legend()
plt.show()
```
